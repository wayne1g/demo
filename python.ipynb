```python
#!/usr/bin/env python3

__author__ = 'wchan'

from pprint import pprint
from jnpr.junos import Device
from datetime import datetime
from jnpr.junos import exception as EzErrors
from lxml import etree
import os

# This could be fqdn and/or IP address
dev_list = [\
            'core1', \
            'core2', \
            '172.17.0.104', \
            ]
dev = [Device(host, user='lab', password='lab123', timeout = 10) for host in dev_list]

# combine the above lists
zip(dev_list, dev)

for item in zip(dev_list, dev):
    print(item[0], item[1])
```

```python
# convert the datetime object to a string and put it as part of the filename
# file_time = current_time.strftime('%Y-%m-%d_%T')
current_time = datetime.now()
file_time = current_time.strftime('%Y-%m-%d_%H_%M_%S')

# create a new directory if it doesn't exist
directory = "./"+file_time
if not os.path.exists(directory):
    os.makedirs(directory)

for item in zip(dev_list, dev):
    dev_name = item[0]
    device = item[1]

    try:
        device.open()
#         device.timeout=10
#         pprint(device.facts)
        print("connect to device - " + dev_name)
        data = device.rpc.get_config(dev_timeout=10, options={'format':'text'})
#         data = device.rpc.get_config(dev_timeout=10, options={'format':'set'})
        config_file = etree.tostring(data, encoding='unicode')
        device.close()
        file_name = dev_name + "-" + file_time + ".txt"
        try:
            with open(directory + "/" + file_name, "w") as text_file:
                text_file.write("{0}". format(config_file))
            text_file.close()
            "config file backup complete - " + dev_name
        except:
            "config file cannot be closed - " + dev_name
#     except EzErrors.ConnectTimeoutError as e:
    except:
        print("ERROR - cannot open device - " + dev_name)
        "config backup FAILED - " + dev_name

# control-enter to run this cell
```
